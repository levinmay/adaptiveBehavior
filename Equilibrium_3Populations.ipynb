## Parameters & Main Functions

```python
N1 = 5000
N2 = 5000
N3 = 5000
N = 15000

I0 = -1 
gamma = 0.1

c12 = 0.005 
c13 = 0.003 
c23 = 0.007 

chat1 = 0.02
chat2 = 0.015
chat3 = 0.01 

alpha1 = 1

alpha21 = 20
alpha22 = 30
alpha23 = 40 

delta = 0
```

```python
def cstar(i,chat, alpha2):
    return chat - ((alpha2 * b0 * i - delta) / (2 * alpha1 * N) )

def eq3_1(i1, i2, i3):
    S1 = N1 - i1
    return i1 + (b0 * cstar(i1+i2+i3,chat1, alpha21) * S1 * i1) + (b0 * c12 * S1 * i2) + (b0 * c13 * S1 * i3) - (gamma * i1) 

def eq3_2(i1, i2, i3):
    S2 = N2 - i2
    return i2 + (b0 * cstar(i1+i2+i3,chat2,alpha22) * S2 * i2) + (b0 * c12 * S2 * i1) + (b0 * c23 * S2 * i3) - (gamma * i2) 

def eq3_3(i1, i2, i3):
    S3 = N3 - i3
    return i3 + (b0 * cstar(i1+i2+i3,chat3,alpha23) * S3 * i3) + (b0 * c13 * S3 * i1) + (b0 * c23 * S3 * i2) - (gamma * i3) 

```

```python
def workablePairs(string):
    workable_pairs = []
    all_pairs = []
    for pair in string.split("{"):
        if "VeryThinSpace" not in pair:
            pairsmall = pair.split("->")
            if len(pairsmall) > 3:
                i1 = float(pairsmall[1].split(",")[0])
                i2 = float(pairsmall[2].split(",")[0])
                i3 = float(pairsmall[3].split("}")[0])

                if i1 <= 5000 and i1 > 0 and i2 <=5000 and i2 > 0 and i3 <=5000 and i3 > 0:
                    workable_pairs.append(i1)
                    workable_pairs.append(i2)
                    workable_pairs.append(i3)
                all_pairs.append(i1)
                all_pairs.append(i2)
                all_pairs.append(i3)
    return workable_pairs,all_pairs
```

```python
def runSolve(string):
   # changeVar(b0_given,c12_given)
    
    return workablePairs(string)
   
```
